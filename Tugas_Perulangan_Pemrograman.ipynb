**2. Buatlah algoritma/program yang membaca sebuah nilai integer positif, misalnya N, dan
menjumlahkan (serta menampilkan) semua bilangan kelipatan 5 antara 1 s.d. N**

```python
# Meminta pengguna memasukkan nilai N
N = int(input("Masukkan nilai N: "))

# Inisialisasi variabel untuk menyimpan total jumlah kelipatan 5
total = 0

# Loop untuk mencari dan menambahkan kelipatan 5
for i in range(1, N + 1):
    if i % 5 == 0:
        total += i

print("Jumlah semua bilangan kelipatan 5 antara 1 hingga", N, "yaitu:", total)
```

**3. Buatlah 3 versi program yang menerima masukan sebuah integer positif, misalnya N (asumsi N>0), dan menuliskan lirik lagu Anak Ayam di atas dengan menggunakan perulangan for dan while.**

```python
# Menggunakan perulangan for

def anak_ayam_for(N):
  lirik = ""
  for i in range(N, 0, -1):
    if i > 1:
      lirik += f"Anak ayam turunlah {i}\nMati satu tinggallah {i-1}\n"
    else:
      lirik += f"Anak ayam turunlah {i}\nMati satu tinggal induknya\n"
  return lirik

N = 5
print(anak_ayam_for(N))
```

```python
# Menggunakan perulangan while

def anak_ayam_while(N):
  lirik = ""
  i = N
  while i > 0:
    if i > 1:
      lirik += f"Anak ayam turunlah {i}\nMati satu tinggallah {i-1}\n"
    else:
      lirik += f"Anak ayam turunlah {i}\nMati satu tinggal induknya\n"
    i -= 1
  return lirik

N = 5
print(anak_ayam_while(N))
```

```python
# Versi 3: Menggunakan fungsi rekursif

def anak_ayam(N):
  if N > 1:
    return f"Anak ayam turunlah {N}\nMati satu tinggallah {N-1}\n" + anak_ayam(N-1)
  else:
    return f"Anak ayam turunlah {N}\nMati satu tinggal induknya\n"

N = 5
print(anak_ayam(N))
```
